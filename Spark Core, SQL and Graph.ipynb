```python
# a)	Save the dataset as a DataFrame, and print the schema.
from pyspark.sql import SQLContext,Row
sqlContext = SQLContext(sc)
TweetsData  = sqlContext.jsonFile("file:///home/cloudera/anurag/tweets.json")
```

```python
TweetsData.printSchema()
```

```python
#b)	Get all of the tweets made by a user (any user would work. We should be able to replace user names to get tweets by that particular user).
```

```python
TweetsData.registerTempTable("newTweets")
uservar = "Daniel Beer"
def getTweetsByUser(uservar):
    res = sqlContext.sql("select text as tweets from newTweets where user='"+uservar+"'")
    return res


results  = getTweetsByUser(uservar) 
results.show()




```

```python
#c)	Find count of all tweets by each user user.

```

```python
sqlContext.sql("select count(*) as TotalTweets ,user as UserName from newTweets group by user order by TotalTweets desc").show()
```

```python
# d)	Get a list of all of the people who are mentioned in tweets.

```

```python
peopleTweets = sqlContext.sql("select text as Tweets from newTweets where text like '%@%'")
```

```python
peopleTweets.show()
```

```python
#below code is erroring out,please suggest why




# peopleRDD  =peopleTweets.rdd
# peopleRDD1 = peopleRDD.map(lambda x:x.split(" "))
```

```python
peopleRDD1 = TweetsData.map(lambda x:x.text.split(" "))
Users = []
for p in peopleRDD1.collect():
    for w in p:
        if w.startswith("@"):
            Users.append(w)
Users
```

```python
# e)	Count the number of time each person is mentioned in the entire dataset of tweets.

```

```python
PeopleData = sc.parallelize(Users)
```

```python
row = Row("person")
PeopleDF = PeopleData.map(row).toDF()
```

```python
PeopleDF.registerTempTable("peopleCount")
```

```python
sqlContext.sql("select count(*) as occurence, person as user from peopleCount group by person order by occurence desc").show()
```

```python
#f)	Give top 50 users who are mentioned the most.

sqlContext.sql("select count(*) as occurence, person as user from peopleCount group by person  order by occurence desc limit 50 ").show()
```

```python
# g)	Get a list of all hashtags mentioned in the dataset.

hashData =  TweetsData.map(lambda x:x.text.split(" "))
hashlist = []
for h in hashData.collect():
    for t in h:
        if t.startswith("#"):
            hashlist.append(t)

hashlist
            
```

```python
# h)	Find how many times each hashtag is mentioned in the dataset.

```

```python
hashData = sc.parallelize(hashlist)
row = Row("hashTag")
hashtags = hashData.map(row).toDF()
```

```python
hashtags.registerTempTable("hashtagsseperated")
```

```python
sqlContext.sql("select count(*) as occurences, hashTag from hashtagsseperated group by hashTag order by occurences desc").show()
```

```python
# Q3-
# a)	Import the data and create a graph using GraphFrames (Hint: Your graph will have nodes and edges.
# Nodes here would be individual stations so id field would be name field in station.csv file.
# Edges would have src and dst so it would Start Station and End Station fields in trip.csv file respectively.
# You can make use of other fields as properties of nodes and edges). 

```

```python
from graphframes import *
edges = sqlContext.read.format("com.databricks.spark.csv").options(header="true").load("file:///home/cloudera/anurag/trip.csv")
Vertices = sqlContext.read.format("com.databricks.spark.csv").options(header="true").load("file:///home/cloudera/anurag/station.csv")
```

```python

edges=edges.withColumnRenamed("Start Station","src")
edges=edges.withColumnRenamed("End Station","dst")
edges.show()
```

```python
Vertices  =Vertices.withColumnRenamed("name","id")
Vertices.show()
```

```python
stationGraphFrame = GraphFrame(Vertices,edges)
```

```python
# b)	Find out number of incoming connections and outgoing connections for each node and print the top 10 nodes.

```

```python
stationGraphFrame.inDegrees.sort("inDegree",ascending = False).limit(10).show()
```

```python
stationGraphFrame.outDegrees.sort("outDegree",ascending = False).limit(10).show()
```

```python
#c)	Find out which are the most common direct routes that people take and print top 10.

```

```python
directRoutes = stationGraphFrame.find ("(src)-[R1]->(dst)")
```

```python
directRoutes.groupBy("src","dst").count().sort("count",ascending=False).limit(10).show()
```
